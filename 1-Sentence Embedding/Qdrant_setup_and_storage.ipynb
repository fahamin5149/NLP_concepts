## Define Storage location

## On-disk storage

Instead of storing everything in memory, you can configure Qdrant to persist data on disk. This allows you to store larger datasets that don't fit in RAM. To use on-disk storage, specify a path when initializing the QdrantClient:

```python
qdrant = QdrantClient(path="/path/to/qdrant/data")
```

## Qdrant server mode

Rather than running Qdrant in local mode, you can use the Qdrant server. This is a production-ready service that supports concurrent access and persistence. To use the server, specify the URL when initializing the client:

```python
qdrant = QdrantClient(url="http://qdrant.example.com:6333")
```

## Qdrant Cloud

For a fully managed Qdrant service, you can use Qdrant Cloud. This provides a hosted Qdrant cluster that you can connect to using a URL and API key:

```python
qdrant = QdrantClient(
    url="https://qdrant.cloud.example.com",
    api_key="your-api-key"
)
```

````python
qdrant= QdrantClient(":memory:")
````

## Create a collection

### `client.create_collection`

This is a method call to create a new collection in the Qdrant database. A collection is a grouping of vectors (data points) that share a common structure and can be queried together.

### Parameters

1. **`collection_name`**:
   - This is the name of the collection you want to create. In your example, the collection is named `"my_books"`. Choose a name that reflects the data it will store.

2. **`vectors_config`**:
   - This parameter defines the configuration for the vectors that will be stored in the collection. It specifies the characteristics of the vectors, including their size and the distance metric used for similarity calculations.

### `models.VectorParams`

This is a configuration object that specifies the parameters for the vectors:

- **`size`**:
  - This should match the dimensionality of the embeddings produced by your encoder model. The method `encoder.get_sentence_embedding_dimension()` retrieves the size of the embedding vectors. Ensure that this matches the output size of the model you are using.

- **`distance`**:
  - This defines the distance metric used for comparing vectors. In your example, `models.Distance.COSINE` indicates that the cosine similarity will be used to measure the similarity between vectors. Other options may include Euclidean distance or inner product, depending on your use case.

````python
qdrant.create_collection(
    collection_name="my_books",
    vectors_config= models.VectorParams(
        size= encoder.get_sentence_embedding_dimension(),
        distance= models.Distance.COSINE,
    )
)
````

### `qdrant.upload_points`

This is a method call to upload a list of points (vectors) to the specified collection in Qdrant.

### Parameters

1. **`collection_name`**:
   - This is the name of the collection to which you want to upload the points. In your example, the collection is named `"my_books"`.

2. **`points`**:
   - This is a list of `PointStruct` objects, each representing a single point (vector) to be uploaded to the collection.

### `models.PointStruct`

This is a data structure that defines a single point (vector) to be uploaded. It has the following properties:

- **`id`**:
  - This is a unique identifier for the point. In your example, it is set to the index of the document in the `documents` list.

- **`vector`**:
  - This is the vector (embedding) associated with the point. In your example, it is obtained by encoding the `"description"` field of each document using the `encoder.encode` method. The resulting vector is converted to a list using `tolist()`.

- **`payload`**:
  - This is an optional field that allows you to store additional data associated with the point. In your example, the entire `doc` dictionary is used as the payload.

````python
qdrant.upload_points(
    collection_name="my_books",
    points= [
        models.PointStruct(
          id=idx,
          vector=encoder.encode(doc["description"]).tolist(),
          payload=doc
        )
    for idx, doc in enumerate(documents)
    ],
)
````

## Ask the engine a question


````python
hits= qdrant.search(
    query_vector=encoder.encode("science fiction").tolist(),
    limit=3,
    collection_name="my_books",
)

for hit in hits:
  print(hit.payload,"Score", hit.score)
````

### `qdrant.search`

This is a method call to perform a search query in the specified collection. It returns a list of `SearchResult` objects representing the most similar points to the query vector.

### Parameters

1. **`query_vector`**:
   - This is the vector (embedding) used as the query. In your example, it is obtained by encoding the phrase "science fiction" using the `encoder.encode` method and converting it to a list.

2. **`limit`**:
   - This specifies the maximum number of results to return. In your example, it is set to `3`, meaning it will return the top 3 most similar points.

3. **`collection_name`**:
   - This is the name of the collection to search within. In your example, the collection is named `"my_books"`.

4. **`query_filter`**:
   - This is an optional parameter that allows you to apply filters to the search results. In your example, a filter is applied using `models.Filter`.

### `models.Filter`

This is a data structure that defines the filter conditions for the search query. It has the following properties:

- **`must`**:
  - This is a list of `FieldCondition` objects that define the mandatory conditions for the filter. In your example, there is one condition that filters documents where the `"year"` field is greater than or equal to `2000`.

### `models.FieldCondition`

This is a data structure that defines a single condition for a field in the payload data. It has the following properties:

- **`key`**:
  - This is the name of the field to filter on. In your example, it is set to `"year"`.

- **`range`**:
  - This defines the range condition for the field. In your example, it is set to `models.Range(gte=2000)`, which means the field value must be greater than or equal to `2000`.

````python
hits= qdrant.search(
    query_vector=encoder.encode("science fiction").tolist(),
    limit=3,
    collection_name="my_books",
    query_filter=models.Filter(
        must=[
            models.FieldCondition(
                key="year",
                range=models.Range(
                    gte=2000,
                )
            )
        ]
    )
)

for hit in hits:
  print(hit.payload,"Score", hit.score)
````
